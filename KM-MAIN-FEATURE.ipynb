```python
import pandas as pd
```

```python
# Mapping lets you format an entire DataFrame
file = "Data/duke_energy_2010.csv"
file_df = pd.read_csv(file)
file_df.head()
```

```python
list(file_df.columns)
```

```python
energy_subset=file_df[["SAMPLE_USAGE","SAMPLE_PREMISE_ZIP"]]

energy_subset_copy = energy_subset.copy()
energy_subset_copy["zipcode"]=energy_subset["SAMPLE_PREMISE_ZIP"].astype(str)
energy_subset_copy.head(15)
```

```python
energy_subset_copy2 = energy_subset_copy.copy()

for row in range(0,len(energy_subset_copy["zipcode"].astype(str))):
    if len(energy_subset_copy["zipcode"].astype(str)) > 5:
       energy_subset_copy2["zipcode"][row] = energy_subset_copy["zipcode"][row][:5]
    else:
       energy_subset_copy2["zipcode"][row] = energy_subset_copy["zipcode"][row]        
```
