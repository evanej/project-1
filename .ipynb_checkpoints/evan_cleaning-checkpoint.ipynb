```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import linregress


# File to Load (Remember to change these)
energy_data = "Data files/duke_energy_2010.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data

households = pd.read_csv("Data/duke_energy_2010.csv")
households.head()
```

```python
# Create test data set with values, attitudes, beliefs. 
attitudes = households[['green', 'payattn', 'considerproduct', 'veryhotdays', 'firstone' , \
                        'conscientiousrecycle', 'reduceenergy', 'realeffort', 'willingchanges', \
                        'highenergyproblem', 'ccreal', 'scarce', 'solar', 'wind', 'muchasican', 'futurepower', 'phev']]

attitudes.head()
```

```python
x_axis = attitudes[['highenergyproblem']]
y_axis = attitudes[['ccreal']]

plt.title("RELATIONSHIP BETWEEN CONCERN ABOUT ENERGY SUPPLY AND CLIMATE CHANGE")
plt.xlabel("Climate Concern")
plt.ylabel("Energy Supply Concern")

(slope, intercept) = linregress(x_axis, y_axis)
fit = slope * x_axis + intercept

fig, ax = plt.subplots()

fig.suptitle("RELATIONSHIP BETWEEN ENERGY SUPPLY CONCERN AND CLIMATE", fontsize=12, fontweight="bold")

ax.set_xlim(0, 10)
ax.set_ylim(0, 10)

ax.set_xlabel("Climate Concern")
ax.set_ylabel("Energy Supply Concern")

ax.plot(x_axis, fake, linewidth=0, marker='o')
ax.plot(x_axis, fit, 'b--')

plt.show()
```
